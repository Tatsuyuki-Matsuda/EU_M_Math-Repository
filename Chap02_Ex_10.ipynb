```python
#Exercise 2-10
import numpy as np
import matplotlib.pyplot as plt

def function(x):
  return (5*x + 3)

x = np.arange(-10, 10)
plt.figure(figsize =(10, 6))
plt.plot(x, function(x))
plt.grid(True)
```
