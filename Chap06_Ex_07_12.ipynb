```python
#Ex-07
import requests, zipfile
from io import StringIO
import io

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip'

r = requests.get(url, stream=True)

z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()

import pandas as pd
student_data_math = pd.read_csv('student-mat.csv', sep=';')
student_data_math['age_d'] = student_data_math['age'].map(lambda x: x*2)
student_data_math.head()
```

```python
#Ex-08
absences_bins = [0,1,5,100]

student_data_math_cut_data = pd.cut(student_data_math.absences, absences_bins, right = False)
pd.value_counts(student_data_math_cut_data)
```

```python
#Ex-09
pd.value_counts(pd.qcut(student_data_math.absences, 3))
```

```python
#Ex-10
student_data_math.groupby('school')['G1'].mean()
```

```python
#Ex-11
student_data_math.groupby(['school', 'sex'])['G1', 'G2', 'G3'].mean()
```
