```python
#Hw-01
#MatPlotlibを使い、自分の卒論のデータを用いてグラフを作成せよ。
import pandas as pd
Mz_data = pd.read_csv('6-axis_Force_data_sample.csv',sep=',')
#卒研で測定した機体のヨーイングモーメント（センサーのZ軸方向）のデータを使用
#ラダー舵角15°時のスロットル0％、18％における作用モーメントの比較グラフ

import matplotlib.pyplot as plt
plt.scatter(Mz_data['time'],Mz_data['Mz1'], label='Rudd15Thr00')
plt.scatter(Mz_data['time'],Mz_data['Mz2'], label='Rudd15Thr18')
plt.legend()
plt.xlabel('time [sec]')
plt.ylabel('Mz [Nm]')
```
