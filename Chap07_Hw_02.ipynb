```python
#Hw-02
#PlotlyInteractive
import pandas as pd
Mz_data = pd.read_csv('6-axis_Force_data_sample.csv',sep=',')

import plotly.graph_objects as go

trace1 = go.Scatter(x=Mz_data['time'], y=Mz_data['Mz1'], name = 'Rudd15Thr00')
trace2 = go.Scatter(x=Mz_data['time'], y=Mz_data['Mz2'], name = 'Rudd15Thr18')
 
fig = go.Figure(data=[trace1, trace2])

fig.update_layout(title='Z()')
 
fig.update_xaxes(title='time [sec]')
fig.update_yaxes(title='Mz [Nm]')

fig.show()
```
